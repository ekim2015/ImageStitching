## 3. Extract local neighborhoods around every keypoint

```python
def neighbor_descriptors(img, r, c, width):
    # Data structure to handle descriptors
    
#     print(img.shape)
    
    # since width is 2 * radius + 1, we do (width - 1) // 2 
    radius = math.floor((width - 1) / 2)
    desc_matrix = np.zeros((len(r), (2*radius+1)**2), dtype=np.float32)
                            
    # Extract local neighborhoods
    length = len(r)
    for i in range(length):
        row = r[i]
        col = c[i]
        
        # handle out of index
        if (row < radius): # i.e. for radius of 50, to be in bounds must add 50
            row += radius
        elif (row >= (img.shape[0]) - radius): # checks if within radius of the right side
            row -= radius
            
        if (col < radius):
            col += radius
        elif(col >= (img.shape[1]) - radius):
            col -= radius
        
#         if (row < radius or row >= (img.shape[0] - radius) or col < radius or col >= (img.shape[1] - radius)):
#             continue
#         print((img.shape[1] + 2) - radius)
#         print("Row and column: ", row, col)
        
        # corners
        x_left = col - radius
        x_right = col + radius + 1
        y_top = row - radius
        y_bottom = row + radius + 1
        
#         print(x_left, x_right, y_top, y_bottom)
        
        # get patch
        patch = img[y_top:y_bottom, x_left:x_right]
#         print("Patch shape: ", patch.shape)
        patch = patch.reshape(-1)
#         print(patch)
#         print("Patch length: ", len(patch))
        desc_matrix[i] = patch
    return desc_matrix
    

# r and c are a bunch of coordinates of corners
# use r and c to find patches of the image

descriptors_1 = neighbor_descriptors(img1, r1, c1, 7)
descriptors_2 = neighbor_descriptors(img2, r2, c2, 7)
```

## 4. Compute distances between descriptors

```python
# sift implementations
def neighbor_descriptors(img, r, c, radius):
    circle_num = len(r)
    circle_arr = np.empty((circle_num, 3))
    for i in range(circle_num):
        x = c[i]
        y = r[i]
        circle = [x, y, radius]
        circle_arr[i] = circle
#     print("Circle array: ", circle_arr)
    return find_sift(img, circle_arr, 1.5)
    
descriptors_1 = neighbor_descriptors(img1, r1, c1, 2)
descriptors_2 = neighbor_descriptors(img2, r2, c2, 2)
print(descriptors_1)
print(descriptors_2)
# print(descriptors_1.shape)
```

```python
# use dist2 from utils.py to compute dist between descriptors
distances = dist2(descriptors_1, descriptors_2)
print(distances)
print(distances.shape)

# (737, 727) means 731 rows, 727 columns
```

## 5. Select Matches

```python
def filter_descriptors_by_dist(distances, thresh, c1, c2):
    # for each row, find all of the distances
    index_dict = {"Distance": [], "Indices": []}
    cache1 = []
    cache2 = []
    count = 0
    for i in range(distances.shape[0]): # corresponds to 2nd image 
        for j in range(distances.shape[1]): # corresponds to 1st image
#             print(i)
            if distances[i, j] <= thresh:
                if i in cache1 or j in cache2 or c1[j] < 500 or c2[i] > 600: # ignore extreme points or duplicate indices
                    continue
                index = (i, j)
                index_dict["Distance"].append(distances[i, j]) 
                index_dict["Indices"].append(index)
                count += 1
                cache1.append(i)
                cache2.append(j)
                print(count)
    return index_dict
                
filtered_matches = filter_descriptors_by_dist(distances, 0.15, c1, c2)
print(filtered_matches)
```

## 6. RANSAC

```python
def findHomography(feature_pairs):
    # array of matrices
    matrix_arr = []
        
    # next compute homography using each pair in feature pairs
    for i in range(len(feature_pairs)):
        # in each pair, first element is from second image and second element is from first image
        # x refers to column[idx] and y refers to row[idx]
#         print(feature_pairs[i])
        xi = feature_pairs[i][0]
        yi = feature_pairs[i][1]
        xi_prime = feature_pairs[i][2]
        yi_prime = feature_pairs[i][3]
        
        # verification of points
#         print(xi, yi, xi_prime, yi_prime)
        
        row_matrix = np.matrix([[xi, yi, 1, 0, 0, 0, -xi_prime*xi, -xi_prime*yi, -xi_prime], [0, 0, 0, xi, yi, 1, -yi_prime*xi, -yi_prime*yi, -yi_prime]])
#         print(row_matrix)
        
        matrix_arr.append(row_matrix)
           
    # shape is right
    
    A = np.vstack(matrix_arr[0:4])
    
    u, s, vh = np.linalg.svd(A, full_matrices=True)
    h = vh[8]
    h_arr = h.reshape(3, 3)
    h_arr = h_arr / h_arr[2, 2]
    
    return h_arr
```

```python
def RANSAC(filtered_matches, r1, r2, c1, c2, eps, num_loops):
    # use np.linalg.svd to compute H
    # find one homography per RANSAC loop
        
    # distance / inliers is the residual    
    inliers = [] # keep track of inliers
    inlier_quantities = []
    residuals = []
    homographies = [] # keep track of homographies
    for i in range(num_loops): # number of samples  
        # Select 4 random feature pairs
        feature_pairs = np.empty((4,4))
        src_pts = np.empty((4, 2))
        dst_pts = np.empty((4, 2))
        for i in range(4):
            index = random.choice(filtered_matches['Indices'])
            xi = c1[index[1]]
            yi = r1[index[1]]
            xi_prime = c2[index[0]]
            yi_prime = r2[index[0]]
            pair = np.array([xi, yi, xi_prime, yi_prime])
            src_pts[i] = [xi, yi]
            dst_pts[i] = [xi_prime, yi_prime]
            feature_pairs[i] = pair
        
#       this is the manual implementation  
#         H = findHomography(feature_pairs)
#         homographies.append(H)
        
        # cv version of findHomography
        H1 = cv.findHomography(src_pts, dst_pts)
        homographies.append(H1[0])
#         print(H1[1].shape)
        
        # even with findHomography, the homography is still erroneous
        # 
    
        inliers_arr = []
        total_residual = []
        inliers_count = 0
        
        # get inliers for all potential matches
        
        points = filtered_matches["Indices"]
        for i in range(len(points)):
            x_prime = c2[points[i][0]]
            y_prime = r2[points[i][0]]
            x = c1[points[i][1]]
            y = r1[points[i][1]]
            
            # calculate the matrices
            xy = np.array([x, y, 1])
            xy_t = np.transpose(xy)
            value = np.matmul(H1[0], xy_t)

            x_pred = value[0] / value[2]
            y_pred = value[1] / value[2]
            
            # this distance is the residual
            val1 = (x_pred - x_prime)**2
            val2 = (y_pred - y_prime)**2
            distance = math.sqrt(val1 + val2)
            total_residual.append(distance**2)
            
            if (distance < eps):
                inliers_arr.append([x, y, x_prime, y_prime])
                inliers_count += 1
            
        # count number of inliers for this homography
        inlier_quantities.append(inliers_count)
        inliers.append(inliers_arr)
        residuals.append(np.mean(total_residual))
        
    loc = np.argmax(inlier_quantities)
#     print(loc)
    
#     print(inlier_quantities)
    
    residual_max = residuals[loc]  
    homography_max = homographies[loc]
    inliers_max = inliers[loc]
#     print(inliers_max)

#     print(homographies)
    
    return inliers_max, residual_max, homography_max
```

```python
# report num of inliers and the average residual for the inliers
inliers, residual, H = RANSAC(filtered_matches, r1, r2, c1, c2, 2, 100)
print("number of inliers: " + str(len(inliers)))
print("residual: " + str(residual))
print("best homography: ", H)
# print(inliers)

# display the locations of inlier matches in both images
```

## 7. Warp one image onto the other

```python
H_warp = findHomography(inliers)
print(H_warp)
```

```python
def warp_img(image_left, H):
    # this part of code is referenced from Ajinkya Tejankar
    # get the height and width of the image
    # this function will work with color images
    # so, having third color channel is not a problem
    h_left, w_left = image_left.shape[:2]

    # we want to find where the image corners are going to land
    # so, we create a matrix of four corner points
    C_left = np.array([
        [0, 0     , w_left, w_left],
        [0, h_left, 0     , h_left],
        [1, 1     , 1     , 1     ]
    ])

    # apply the homography to the corner points to get projected corner points
    Cp_left = H @ C_left
    Cp_left = Cp_left / Cp_left[-1, :]

    # find the minimum height and width of the projected corners
    w_min, h_min = Cp_left[:-1].min(axis=1).tolist()
    # we might need to properly floor or ceil the floats to prevent
    # the edge pixels from getting cropped but this works for our needs
    # feel free to fix this
    w_min, h_min = int(np.abs(w_min)), int(np.abs(h_min))
    # what's the final warped image size that can hold the full image?
    warped_image_shape = (h_left + h_min, w_left + w_min)

    # we create a new homography that applies the translation
    # that would be otherwise cropped by the warp function below
    Ht = np.array([
        [1, 0, w_min],
        [0, 1, h_min],
        [0, 0, 1    ]
    ])
    # apply the translation homography so that the image is warped
    # but does not have a negative translation relative to origin
    Hw = Ht @ H
    # may not be strictly necessary but make sure that (3,3) is 1
    Hw = Hw / Hw[-1, -1]

    # use skimage.transform.ProjectiveTransform to create a transform for the homography Hw
    tform = skimage.transform.ProjectiveTransform(Hw)
    # use skimage.transform.warp to apply the transform
    warped_image = skimage.transform.warp(image_left, tform.inverse, output_shape=warped_image_shape)

    return warped_image
```

```python
# best result with warping using my RANSAC homography
warped_img1 = warp_img(img1, H_warp)
print(warped_img1.shape)
plt.imshow(warped_img1, cmap='gray')
```

## 8. Create a new image to hold the panorama

```python
def create_panorama(warped_img1, img2):
    print(warped_img1.size[0])
    img_new = Image.new('L', (warped_img1.size[0] + 260, warped_img1.size[1]))
    print(img2)
    img_new.paste(warped_img1, (0, 0))
    img_new.paste(img2, (1000, 317))
    image_arr = np.array(img_new)
    
    warped_img1_arr = np.array(warped_img1)
    # averaging
    for i in range(image_arr.shape[0] - 1):
        for j in range(image_arr.shape[1] - 1):
            if (image_arr[i, j] == 0):
                continue
            else:
                if (image_arr[i, j] > 0 and warped_img1_arr[i,j] > 0):
                    image_arr[i, j] = (image_arr[i, j] + warped_img1_arr[i,j]) / 2
                else:
                    continue
    
    return image_arr
    

# create a panorama in gray scale first
panorama_gray = create_panorama(warped_img1, image2)
plt.imshow(panorama_gray, cmap='gray')
```

```python
# then create a color image by doing the same for each channel
def create_panorama_rgb(img1, img2):
    img_new = Image.new('RGB', (img1.size[0] + 260, img1.size[1]))
    img_new.paste(img1, (0, 0))
    img_new.paste(img2, (1000, 317))
    
    img_arr = np.array(img_new)
    print(img_arr.shape)

    warped_img1_arr = np.array(img1)
    # now we go for each RGB channel
    for c in range(1, 3):
        for i in range(img_arr.shape[0] - 1):
            for j in range(img_arr.shape[1] - 1):
                if (img_arr[i, j, c] == 0):
                    continue
                else:
                    if (img_arr[i, j, c] > 0 and warped_img1_arr[i,j,c] > 0):
                        img_arr[i, j, c] = (img_arr[i, j, c] + warped_img1_arr[i,j,c]) / 2
                    else:
                        continue    
            
    return img_arr

# for image 1, take the warped image and make it colored
img1_color = warped_img1.convert('RGB')
print(np.array(img1_color).shape)
img2_color = Image.open('./hw2_data/uttower_right.JPG')

# plt.imshow(img2_color)
panorama = create_panorama_rgb(img1_color, img2_color)
plt.imshow(panorama)
```
